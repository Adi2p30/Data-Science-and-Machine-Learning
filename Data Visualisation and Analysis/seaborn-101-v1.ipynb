# Introduction to Seaborne
Seaborn is a really cool Python library that helps you make fancy graphs and charts. It's built on top of Matplotlib, another Python library, and it's super easy to use. Seaborn comes with a bunch of datasets and tools that let you make all sorts of cool graphs, like scatterplots, line plots, and bar charts. It even has heatmaps, which are really useful for visualizing data that changes over time or across different categories.

What's really great about Seaborn is that it works really well with Pandas dataframes, which are basically like tables of data. If you have a cleaned and formatted dataframe, you can use Seaborn to quickly make a graph that shows you what's going on. Seaborn also has features for working with categorical data, which is data that's divided into groups or categories. It can help you visualize distributions and relationships between different variables, too.

One thing that sets Seaborn apart from other visualization libraries is that it looks really nice right out of the box. It has a bunch of different themes and color palettes that you can use to make your graphs look really cool without having to do a lot of work. And even though Seaborn can do some pretty advanced stuff, its API is designed to be really easy to use, so even beginners can get started with it pretty quickly.!

```python
import numpy as np 
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt 
```

* # Importing the Dataset
The tips dataset is an example dataset that is included with the Seaborn data visualization library for Python. It contains information about customers who visited a restaurant, such as the total bill they paid, the amount of tip they gave, and other variables like the day of the week, time of day, and whether they were a smoker or not. This dataset is frequently used to demonstrate the various types of graphs and charts that can be created using Seaborn. For instance, scatterplots can be created to examine the relationship between the total bill and tip amount, or boxplots can be used to compare the distribution of tips between smokers and non-smokers. The tips dataset is an interesting way to explore real-world data related to restaurant dining and an excellent resource for students who want to learn more about data visualization and Seaborn.

```python
tips = sns.load_dataset("tips")
tips.head()
```

# Line Chart
A line chart, also known as a line plot, is a graphical representation of data that displays information as a series of data points connected by straight line segments. Line charts are commonly used to show trends over time, and are useful for visualizing continuous data.

```python
x = tips["total_bill"]
y = tips["tip"]
sns.lineplot(x=x, y=y)
plt.xlabel('Total Bill')
plt.ylabel('Tips')
plt.title('Line Chart Example')
plt.show()
```

# Box Plot
A box plot, also known as a box and whisker plot, is a graphical representation of a dataset that displays the distribution of the data along with the outliers. The box in the plot represents the interquartile range (IQR), which is the range between the 25th and 75th percentiles of the data. The whiskers extend from the box and represent the range of the data within 1.5 times the IQR. Data points outside of this range are considered outliers and are plotted individually.
In Python, the Seaborn library provides an easy way to create box plots. 

```python
data = tips
sns.boxplot(data=data)
plt.title('Box Plot Example')
plt.show()
```

# Rel Plot  
Seaborn's `relplot()` function is a powerful tool for visualizing the relationships between variables. It allows you to create a wide range of plots, including scatter plots, line plots, and categorical plots. The `relplot()` function is designed to be flexible and easy to use, and it allows you to customize many aspects of your plot, such as the size and shape of your markers, the color of your lines, and the labels on your axes.

```python
x = tips["total_bill"]
y = tips["tip"]
sns.relplot(x=x, y=y)
plt.title('Rel Plot Example')
plt.show()
```

# Count Plot 
A count plot is a type of graph that is used to display the count of observations for a given categorical variable. In a count plot, the number of occurrences of each category is represented by a bar of equal width. The height of each bar represents the count of observations in that category.

```python
sns.countplot(x='sex', data=tips, hue='smoker')
plt.title('Count Plot Example')
plt.show()
```

# Swarm Plot 
A swarm plot is a type of categorical scatter plot that displays the relationship between a numeric variable and a categorical variable. It is similar to a strip plot, but instead of overlaying the points, they are adjusted so that they do not overlap. This allows you to see the distribution of data points more clearly and can be useful for visualizing small datasets.

```python
# x = tips.loc[range(0,10),"total_bill"]
# y = tips.loc[range(0,10),"tip"]
sns.swarmplot(x=tips["day"], y=tips["total_bill"])
plt.title('Swarm Plot Example')
plt.show()
```

# Bar Plot 
A bar plot is a visualization that displays data using rectangular bars with heights or lengths proportional to the values they represent. Bar plots are commonly used to show comparisons between different categories of data.

```python
sns.barplot(x=tips["day"], y=tips["total_bill"])
plt.title('Bar Plot Example')
```

# Heat Map 
A heat map is a type of visualization that displays data in a matrix format using colors to represent different values. Heat maps are commonly used to show correlations between variables or to visualize large datasets. In Python, the Seaborn library provides an easy way to create heat maps.

```python
sns.heatmap(data=tips.corr(),annot=True, cmap="YlGnBu")
plt.title('Heat Map Example')
plt.show()
```

# Point Plot 
A point plot is a visualization that shows point estimates and confidence Intervals as a function of one or more categorical variables. Point plots can be used to show differences between groups and to visualize trends over time.

```python
sns.pointplot(x="day", y="total_bill", data=tips)
plt.title('Point Plot Example')
plt.show()
```

# Strip Plots 
A strip plot is a type of visualization that displays the distribution of data points as a set of vertical or horizontal lines, where each line represents an individual data point. Strip plots are useful for visualizing the spread of data and identifying outliers.

```python
sns.stripplot(x=tips["day"], y=tips["total_bill"])
plt.title('Strip Plot Example')
plt.show()
```

# Pair Grids
A pair grid is a type of visualization that allows you to create a grid of plots for pairwise relationships in your data. Each plot in the grid displays the relationship between two variables, with one variable plotted on the x-axis and the other variable plotted on the y-axis. Pair grids are useful for identifying patterns and correlations in large datasets.

```python
sns.pairplot(data=tips, hue="tip")
```

# Cat Plots 
A cat plot (short for categorical plot) is a type of visualization in Seaborn that allows you to plot the relationship between a numerical variable and one or more categorical variables. This is done by creating separate subplots for each category and displaying the distribution of the numerical variable within each category. Cat plots are useful for visualizing the distribution of data across different categories and identifying any differences or similarities between them.

```python
sns.catplot(x="day", y="total_bill", hue="sex", data=tips, kind="swarm")
plt.title("Total Bill by Day and Gender")
plt.show()
```

# Boxen Plot
A boxen plot, also known as a letter-value plot or a beeswarm plot, is a type of visualization in Seaborn that allows you to plot the distribution of a numeric variable across different categories. The plot is similar to a box plot, but it provides more detailed information about the distribution of the data, especially for larger datasets. In a boxen plot, the boxes are drawn using quartiles, and the outliers are displayed as individual points.


```python
sns.boxenplot(x="day", y="total_bill", hue="sex", data=tips)
plt.title("Total Bill by Day and Gender")
plt.show()
```

# Violin Plot 
A violin plot is a type of visualization in Seaborn that allows you to plot the 
distribution of a numeric variable across different categories. It is similar to a box plot, but it also displays the shape and density of the distribution. In a violin plot, each category is represented by a vertical "violin" shape that shows the density of the data at different values along the y-axis.

```python
sns.violinplot(x="day", y="total_bill", hue="sex", data=tips)
plt.title("Total Bill by Day and Gender")
plt.show()
```

# KDE Plot 
A KDE (Kernel Density Estimation) plot is a type of visualization in Seaborn that allows you to visualize the distribution of a numeric variable. It estimates the probability density function of a continuous variable, giving you a smooth curve that represents the distribution of the data.


```python
sns.kdeplot(x="total_bill", data=tips)
plt.title("Distribution of Total Bill")
plt.show()
```

# Dist Plot 
A distplot (short for "distribution plot") is a type of visualization in Seaborn that allows you to plot a univariate distribution of observations. It combines a histogram of the data with a kernel density estimate (KDE) curve, giving you a comprehensive view of the distribution of your data.!

```python
sns.distplot(tips["total_bill"])
plt.title("Distribution of Total Bill")
plt.show()
```

# Scatter Plot 
A scatter plot is a type of plot in Seaborn that allows you to visualize the relationship between two continuous variables in your data. It can be used to identify patterns, trends, and outliers in your data.

```python
sns.scatterplot(data=tips, x="total_bill", y="tip")
plt.title("Total Bill vs. Tip")
plt.xlabel("Total Bill")
plt.ylabel("Tip")
plt.show() 
```

# Pair Plot 
A Pair Plot in Seaborn is a type of plot that displays pairwise relationships between different variables in a dataset. It is useful for identifying patterns and correlations between different variables in your data.

```python
sns.pairplot(tips)
plt.show()
```

# Facet Grid 
Facet Grid in Seaborn is a type of plot that allows you to create multiple plots side by side, based on the values of one or more variables in your data. It is useful for visualizing relationships between multiple variables in your data.

```python
g = sns.FacetGrid(tips, col="time", row="sex")
g.map(sns.scatterplot, "total_bill", "tip")
plt.show()
```

# Regression Plot
A Regression Plot in Seaborn is a type of plot that displays the relationship between two variables, along with a linear regression line that indicates the overall trend of the data.

```python
sns.regplot(x="total_bill", y="tip", data=tips)
plt.show()
```

# Joint Plot
A Joint Plot in Seaborn is a type of plot that displays the relationship between two variables, along with the distribution of each variable separately. It is useful for visualizing bivariate relationships and identifying patterns in your data.

```python
sns.jointplot(x="total_bill", y="tip", data=tips)
plt.show()
```
